```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import *
porterStemmer = PorterStemmer()
```

Here by we start cleaning up the data for further analysis. Firt step will be to simplify the column weather conditions which contains longer text defining the weather at time of accident. We need to convert the text to simple categorical data

```python
import string

def text_process(text):
    """
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Stemming the words to the root
    3. Returns a list of the cleaned text
    """
    # Check characters to see if they are in punctuation
    nopunc = [char for char in text if char not in string.punctuation]

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
    
    # Now just remove any stopwords
    return [porterStemmer.stem(word.strip()) for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

```python
from collections import Counter
```

```python
c = Counter()
for xs in accidentData['Weather_Condition'].dropna():
    for x in text_process(xs):
        c[x]+=1
```

```python
weather_keywords = ['rain', 'snow', 'thunderstorm','fog', 'wind','cloudy', 'dust', 'overcast','clear' ]
```

```python
weather_dict = {'rain': ['scatter','drizzle', 'wintri', 'mix', 'shower', 'precipit'],
                'snow':['ice', 'hail', 'sleet','pellet','freez'],
                'thunderstorm': ['tstorm','thunder'],
                'fog' : ['haze', 'smoke', 'mist'],
                'wind' : ['windi', 'blow', 'whirlwind', 'tornado', 'whirl', 'drift', 'funnel', 'squall'],
                'cloudy': ['cloud','ash','volcan', 'cloudi'],
                'dust' : ['grain','sand'],
                'overcast':[],
                'clear':[]}
def findSimilarWord(word):
    if word in weather_dict.keys():
        return word
    else:
        for x, y in weather_dict.items():
            if word in y:
                return x
        return ""

            
    
```

```python
findSimilarWord('cloudi')
```

```python
def simplifyWeatherCondition(text):
    bagofWords = text_process(text)
    for word in bagofWords:
        keyword = findSimilarWord(word)
        if (keyword!= ''):
            break
    return keyword
```
