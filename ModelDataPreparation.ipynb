```python
import pandas as pd
import random
from tqdm.notebook import tqdm
```

```python
accidentData = pd.read_csv("finalDatawithClustering.csv")
```

```python
accidentData.shape
```

Creating Non Accident data for training the model

```python
#create all data to be fed to logistic regression
modelData = pd.DataFrame(columns=['Start_Lat', 'Start_Lng','Cluster','Weekday','Hour', 'Date', 
                                    'Month','weather_keyword', 'Sunrise_Sunset','Traffic_Signal',
                                   'Station','Junction','Crossing','Temperature(F)', 'Wind_Speed(mph)','Wind_Direction',
                                   'Pressure(in)','Visibility(mi)','Humidity(%)','Accident'])

#for each of the accidents, create 3 other data points
for index, row in accidentData.iterrows():
    
    #add itself
    latitude = row['Start_Lat']
    longitude = row['Start_Lng']
    nclust = row['Cluster']
    dweek = row['Weekday']
    nhr = row['Hour']
    nday = row['Date']
    month = row['Month']
    wkeyword = row['weather_keyword']
    sunriseset = row['Sunrise_Sunset']
    tsignal = row['Traffic_Signal']
    station = row['Station']
    junction = row['Junction']
    crossing = row['Crossing']
    temp = row['Temperature(F)']
    windspeed = row['Wind_Speed(mph)']
    winddir = row['Wind_Direction']
    vis = row['Visibility(mi)']
    hum = row['Humidity(%)']
    pres = row['Pressure(in)']
    accident = 1
    rowdata = pd.DataFrame([[latitude,longitude,nclust,dweek,nhr,nday,month,wkeyword,sunriseset,tsignal,station,junction,
                          crossing,temp,windspeed,winddir,pres,vis,hum,accident]], columns=['Start_Lat', 'Start_Lng','Cluster','Weekday','Hour', 'Date', 
                                                                                            'Month','weather_keyword', 'Sunrise_Sunset','Traffic_Signal',
                                                                                            'Station','Junction','Crossing','Temperature(F)', 'Wind_Speed(mph)','Wind_Direction',
                                                                                            'Pressure(in)','Visibility(mi)','Humidity(%)','Accident'])
    modelData = modelData.append(rowdata,ignore_index=True)
    
    i=0
    while(i<2):
        #Finding random latitude and longitude from the same county where accident has not taken place.                                                                                    
        county = row['County']
        numofAccInCounty = len(accidentData[accidentData['County'] == county]) 
        randomNumber = random.randint(0, numofAccInCounty-1)                  
        replacementrow = accidentData[accidentData['County'] == county].iloc[randomNumber]
        latitude = replacementrow['Start_Lat']                                                                                  
        longitude = replacementrow['Start_Lng'] 
        nclust = replacementrow['Cluster']
                                                                                            
        dweek = row['Weekday']
        nhr = row['Hour']
        nday = row['Date']
        month = row['Month']
        wkeyword = replacementrow['weather_keyword']
        sunriseset = replacementrow['Sunrise_Sunset']
        tsignal = replacementrow['Traffic_Signal']
        station = replacementrow['Station']
        junction = replacementrow['Junction']
        crossing = replacementrow['Crossing']
        temp = replacementrow['Temperature(F)']
        windspeed = replacementrow['Wind_Speed(mph)']
        winddir = replacementrow['Wind_Direction']
        vis = replacementrow['Visibility(mi)']
        hum = replacementrow['Humidity(%)']
        pres = replacementrow['Pressure(in)']
        accident = 0                                                                      
  
        #whether such record exist first    
        isacc = modelData.loc[(modelData['Cluster'] == nclust) & (modelData['Date'] == nday) & (modelData['Hour'] == nhr)]
        i+=1
        if isacc.empty:
            rowdata = pd.DataFrame([[latitude,longitude,nclust,dweek,nhr,nday,month,wkeyword,sunriseset,tsignal,station,junction,
                          crossing,temp,windspeed,winddir,pres,vis,hum,accident]], columns=['Start_Lat', 'Start_Lng','Cluster','Weekday','Hour', 'Date', 
                                                                                            'Month','weather_keyword', 'Sunrise_Sunset','Traffic_Signal',
                                                                                            'Station','Junction','Crossing','Temperature(F)', 'Wind_Speed(mph)','Wind_Direction',
                                                                                            'Pressure(in)','Visibility(mi)','Humidity(%)','Accident'])
            modelData = modelData.append(rowdata,ignore_index=True)
        else:
            print([nclust,nday,nhr])
            print("Accident found!")
    
```
